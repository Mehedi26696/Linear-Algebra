<center><h1 style="color:green">Matrix Diagonalization</center>


Matrix diagonalization is the process of converting a square matrix $ A $ into its diagonal form $ D $, using the equation:

$$ D = P^{-1} A P $$

where $ P $ is the modal matrix formed by the eigenvectors of $ A $, and $ D $ is the diagonal matrix containing the eigenvalues of $ A $.

---

## Steps for Diagonalization

### Step 1: Define $ D $
The diagonal matrix $ D $ is defined as:

$$
D = \begin{bmatrix}
\lambda_1 & 0 & 0 \\
0 & \lambda_2 & 0 \\
0 & 0 & \lambda_3
\end{bmatrix}
$$

where $ \lambda_1, \lambda_2, \lambda_3 $ are the eigenvalues of $ A $.

### Step 2: Find Eigenvalues $ \lambda $
Solve the characteristic equation:

$$ \text{det}(A - \lambda I) = 0 $$

to compute the eigenvalues of $ A $.

### Step 3: Find Eigenvectors $ X_i $
For each eigenvalue $ \lambda_i $, solve the equation:

$$ (A - \lambda_i I) X_i = 0 $$

to find the corresponding eigenvector $ X_i $.

### Step 4: Construct the Modal Matrix $ P $
The modal matrix \( P \) is formed by arranging the eigenvectors $ X_1, X_2, X_3 $ as columns:

$$ P = \begin{bmatrix} X_1 & X_2 & X_3 \end{bmatrix} $$

### Step 5: Find $ P^{-1} $
Compute the inverse of the modal matrix $ P $.

### Step 6: Compute $ D $
Finally, compute the diagonal matrix $ D $ using the formula:

$$ D = P^{-1} A P $$

---

## Example: Diagonalize the Matrix $ A $

Given the matrix:

$$
A = \begin{bmatrix}
1 & 0 & -1 \\
1 & 2 & 1 \\
2 & 2 & 3
\end{bmatrix}
$$

### Step 1: Find Eigenvalues $ \lambda $
Solve $ \text{det}(A - \lambda I) = 0 $:

$$
\text{det}\begin{bmatrix}
1 - \lambda & 0 & -1 \\
1 & 2 - \lambda & 1 \\
2 & 2 & 3 - \lambda
\end{bmatrix} = 0
$$

The eigenvalues are:

$$ \lambda_1 = 1, \lambda_2 = 2, \lambda_3 = 3 $$

### Step 2: Find Eigenvectors $ X_i $
For each eigenvalue $ \lambda_i $:

- **For $ \lambda_1 = 1 $:**
  Solve $ (A - \lambda_1 I) X_1 = 0 $:

  $$
  X_1 = \begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix}
  $$

- **For $ \lambda_2 = 2 $:**
  Solve $ (A - \lambda_2 I) X_2 = 0 $:

  $$
  X_2 = \begin{bmatrix} -2 \\ 1 \\ 2 \end{bmatrix}
  $$

- **For $ \lambda_3 = 3 $:**
  Solve $ (A - \lambda_3 I) X_3 = 0 $:

  $$
  X_3 = \begin{bmatrix} 1 \\ -1 \\ -2 \end{bmatrix}
  $$

### Step 3: Construct the Modal Matrix $ P $

The modal matrix $ P $ is:

$$
P = \begin{bmatrix}
1 & -2 & 1 \\
-1 & 1 & -1 \\
0 & 2 & -2
\end{bmatrix}
$$

### Step 4: Find $ P^{-1} $

Compute the inverse of $ P $:

$$
P^{-1} = \frac{1}{2} \begin{bmatrix}
0 & -2 & 1 \\
-2 & -2 & 0 \\
1 & 0 & -1
\end{bmatrix}
$$

### Step 5: Compute $ D $

Using $ D = P^{-1} A P $, the diagonal matrix $ D $ is:

$$
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 2 & 0 \\
0 & 0 & 3
\end{bmatrix}
$$

---

## Key Observations

1. Diagonalization requires the matrix to have enough linearly independent eigenvectors. For an $ n \times n $ matrix, there must be $ n $ eigenvectors.
2. The eigenvalues directly populate the diagonal elements of $ D $.
3. The modal matrix $ P $ plays a crucial role in transitioning $ A $ to $ D $.
